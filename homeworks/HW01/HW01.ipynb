```python
# Импортируем библиотеку NumPy и даём ей стандартное сокращение np
import numpy as np
```

```python
# Создаём массив из обычного Python-списка
# NumPy автоматически преобразует список в массив
a = np.array([1, 2, 3, 4])
a
```

```python
# Создаём массив, заполненный нулями
# zeros принимает форму массива (кол-во строк и столбцов)
z = np.zeros((2, 3))
z
```

```python
# Создаём массив, заполненный единицами
o = np.ones(5)
o
```

```python
# Создаём последовательность чисел с помощью arange
# от 0 до 10 (не включая), шаг — 2
r = np.arange(0, 10, 2)
r
```

```python
# Выводим форму массива и тип данных
print("Форма массива a:", a.shape)   # Количество элементов (и измерений)
print("Тип данных массива a:", a.dtype)  # Тип данных внутри массива
```

```python
# Получение одного элемента массива
# Индексация начинается с нуля
print("Второй элемент массива a:", a[1])
```

```python
# Получение среза массива (подмассива)
# Берём элементы массива r со 2-го по 4-й (не включая последний индекс)
print("Срез массива r (элементы r[1:4]):", r[1:4])
```

```python
# Создаём второй массив такой же длины
b = np.array([10, 20, 30, 40])

# Поэлементное сложение двух массивов
print("Сложение a + b:", a + b)
```

```python
# Умножение каждого элемента массива на число
print("a * 3:", a * 3)
```

```python
# Вычисление средней величины массива
print("Среднее значение массива b:", b.mean())
```
